```python
fig, ax=plt.subplots()
ax.bar(data_merge['WHO Region'],data_merge['Deaths'])
plt.show()
#labels are too long some are still unreadable must roatate them 
```

```python
fig, ax=plt.subplots()
ax.bar(data_merge['WHO Region'],data_merge['Deaths'])
ax.set_xlabel('WHO Regions')
ax.set_xticks(data_merge['WHO Region'])
ax.set_xticklabels(data_merge['WHO Region'],rotation=90)
ax.set_ylabel('Number of Deaths')
plt.show()
#from this we can easily visualize the proportion of deaths in americas compared to the other regions
```

```python
fig, ax=plt.subplots()

ax.bar(data_merge['WHO Region'],data_merge['Recovered'],label='Recovered')
ax.bar(data_merge['WHO Region'],data_merge['Active'],label='Active')
ax.set_xlabel('WHO Regions')
ax.set_xticks(data_merge['WHO Region'])
ax.set_xticklabels(data_merge['WHO Region'],rotation=90)
ax.set_ylabel('Number of Confirmed cases')
ax.legend()
plt.show()
#Want to add more data so as to show comparison  between those recovered and those R.I.P. abd those still active =total confirmed cases
```

```python
WHO_column= data_merge['WHO Region']
regions=WHO_column.unique()
print(regions)
```

```python
fig, ax=plt.subplots()
for region in regions:
    region_df = data_merge[data_merge['WHO Region']==region]
    ax.bar(region, region_df['Doses administered per 100 people'].mean(),yerr=region_df['Doses administered per 100 people'].std())
ax.set_xlabel('Regions')
ax.set_ylabel("Doses administered per 100 people")
ax.set_xticklabels(regions,rotation=45)
plt.show()
#Here we iterate over the values of regions setting region as the loop variable
#in each iteration the rows where the WHO Region column is equal to regions is extracted
#a bar is added to the provided ax object, labeled with the region names, with the mean of the "Doses administered per 100 people" column as its height, and the standard deviation as a y-axis error bar.
```
